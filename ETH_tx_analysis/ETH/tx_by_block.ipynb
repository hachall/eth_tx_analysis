```python
import csv
import time 
import datetime
import requests
import pandas as pd

verbose = True
API_KEY = "BRD3W8EJU9UNN3QWQ8G11YNKN5NT5JGAC8"
```

```python
#Get block range from dates

start = datetime.date(2019, 9, 1)
unix_start = int(time.mktime(start.timetuple()))
end = datetime.date(2019, 9, 4)
unix_end = int(time.mktime(end.timetuple()))

s = requests.get(f"https://api.etherscan.io/api?module=block&action=getblocknobytime&timestamp={unix_start}&closest=before&apikey={API_KEY}")
start_block = int(s.json()['result'])
e = requests.get(f"https://api.etherscan.io/api?module=block&action=getblocknobytime&timestamp={unix_end}&closest=before&apikey={API_KEY}")
end_block = int(e.json()['result'])
```

```python
total_blocks = end_block - start_block
transactions = []

i = start_block
while (i <= end_block):
    print(f"Fetching block {i} - {i - start_block} of {total_blocks}")
    
    try:  
        x = requests.get(f"https://api.etherscan.io/api?module=proxy&action=eth_getBlockByNumber&tag={hex(i + 1)}&boolean=true&apikey={API_KEY}", timeout=None)
        block = x.json()

        if 'status' not in block.keys():
            transactions += block['result']['transactions']
            print(f"{len(block['result']['transactions'])} transactions found")
        else:
            time.sleep(2)
            
    except: 
        pass
            
    i +=1
    
    
    
df_tx = pd.DataFrame(transactions)
df_tx.to_csv('./2019-9-1-to-2019-9-4-transactions.csv')
print(f'Success, {len(transactions)} found')
```
