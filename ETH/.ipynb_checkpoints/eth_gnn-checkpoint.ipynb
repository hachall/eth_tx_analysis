```python
class Node():
    def __init__(self, typ, address):
        self.typ = typ
        self.address = address
        self.txs_in = []
        self.txs_out = []
        self.label = ''
        self.name = ''
        
    def add_tx_in(self, tx):
        self.txs_in.append(tx)
        
    def add_tx_out(self, tx):
        self.txs_out.append(tx)
        
    def rand_int_feature_vector(self ):
        return [randint(0,100) for x in list(range(15))]
    
    def agg_txs_amounts(self, txs):
        amounts = []
        for tx in txs:
            amounts.append(float(tx.value))
        return amounts
    
    def txs_time_series(self, txs): #calculate mins between transactions
        times = []
        
        for i, tx in enumerate(txs):
       
            if i == 0:
                continue
                
            d = tx.timestamp
            d_prev = txs[i - 1].timestamp
            diff = d - d_prev
            duration_in_s = diff.total_seconds()
            mins = divmod(duration_in_s, 60)[0]
            times.append(mins)
            
        return times
            
    
    def stats(self, l):
        min_val = 0 if len(l) == 0 else min(l)
        max_val = 0 if len(l) == 0 else max(l)
        avg = 0 if len(l) == 0 else mean(l)
        med = 0 if len(l) == 0 else median(l)
        std = 0 if len(l) in [0, 1] else stdev(l)
        
        return len(l), min_val, max_val, avg, med, std
    
    def tx_type_count(self, txs, tx_type):
        count = 0
        for tx in txs:
            if tx.typ == tx_type:
                count +=1
                
        return count
    
    def num_unique_addr(self, txs, tx_type):
        contracts = set()
        for tx in txs:
            if tx.typ == tx_type:

                contracts.add(tx.contract)
            
        return len(contracts)
    
    def unique_to(self):
        to = set()
        for tx in self.txs_out:
            to.add(tx.receiver)
            
        return len(to)
    
    def unique_in(self):
        incoming = set()
        for tx in self.txs_in:
            incoming.add(tx.sender)
            
        return len(incoming)
    
    def feature_vector(self, label_switch=False):
        in_amounts = self.agg_txs_amounts(self.txs_in)
        out_amounts = self.agg_txs_amounts(self.txs_out)
        all_amounts = self.agg_txs_amounts(self.txs_in + self.txs_out)
        
        in_num, in_min, in_max, in_mean, in_med, in_std = self.stats(in_amounts)
        out_num, out_min, out_max, out_mean, out_med, out_std = self.stats(out_amounts)
        comb_num, comb_min, comb_max, comb_mean, comb_med, comb_std = self.stats(all_amounts)

        in_time_series = self.txs_time_series(self.txs_in)
        out_time_series = self.txs_time_series(self.txs_out)
        both = self.txs_in + self.txs_out
        both.sort(key= lambda x: x.timestamp)
        all_time_series = self.txs_time_series(both)
        
        _, in_ts_min, in_ts_max, in_ts_mean, in_ts_med, in_ts_std = self.stats(in_time_series)
        _, out_ts_min, out_ts_max, out_ts_mean, out_ts_med, out_ts_std = self.stats(out_time_series)        
        _, all_ts_min, all_ts_max, all_ts_mean, all_ts_med, all_ts_std = self.stats(all_time_series)   
        
        token_tx_in = self.tx_type_count(self.txs_in, 'Token')
        token_tx_out = self.tx_type_count(self.txs_out, 'Token')
        uniq_tokens_in = self.num_unique_addr(self.txs_in, 'Token')
        uniq_tokens_out = self.num_unique_addr(self.txs_out, 'Token')
        
        internal_tx_in = self.tx_type_count(self.txs_in, 'Internal')
        internal_tx_out = self.tx_type_count(self.txs_out, 'Internal')

        ci_tx_in = self.tx_type_count(self.txs_in, 'Contract Invocation')
        ci_tx_out = self.tx_type_count(self.txs_out, 'Contract Invocation')
        
#         cc_tx_in = self.tx_type_count(self.txs_in, 'Contract Creation')
        cc_tx_out = self.tx_type_count(self.txs_out, 'Contract Creation')
        
        t_tx_in = self.tx_type_count(self.txs_in, 'Transfer')
        t_tx_out = self.tx_type_count(self.txs_out, 'Transfer')
        
        type_vec = self.one_hot_label(['Wallet', 'Smart Contract', 'Exchange'], False) if label_switch else []
        
#         token_tx_in = 0 if comb_num == 0 else token_tx_in / comb_num 
#         token_tx_out = 0 if comb_num == 0 else token_tx_out / comb_num 
#         internal_tx_in = 0 if comb_num == 0 else internal_tx_in / comb_num 
#         internal_tx_out = 0 if comb_num == 0 else internal_tx_out/ comb_num 
#         ci_tx_in = 0 if comb_num == 0 else  ci_tx_in / comb_num 
#         ci_tx_out = 0 if comb_num == 0 else ci_tx_out / comb_num 
#         cc_tx_out = 0 if comb_num == 0 else cc_tx_out / comb_num 
#         t_tx_in = 0 if comb_num == 0 else t_tx_in / comb_num 
#         t_tx_out = 0 if comb_num == 0 else t_tx_out / comb_num 
        
#         in_num = 0 if comb_num == 0 else in_num / comb_num 
#         out_num = 0 if comb_num == 0 else out_num / comb_num
        
        unique_in = self.unique_in()
#         unique_in = 0 if comb_num == 0 else unique_in / comb_num 
        unique_to = self.unique_to()
#         unique_to = 0 if comb_num == 0 else unique_to / comb_num 

        ls =  [
                in_num, in_min, in_max, in_mean, in_med, in_std,
                out_num, out_min, out_max, out_mean, out_med, out_std,
                comb_num, comb_min, comb_max, comb_mean, comb_med, comb_std,
                in_ts_min, in_ts_max, in_ts_mean, in_ts_med, in_ts_std,
                out_ts_min, out_ts_max, out_ts_mean, out_ts_med, out_ts_std,
                all_ts_min, all_ts_max, all_ts_mean, all_ts_med, all_ts_std,
            
                token_tx_in, token_tx_out, uniq_tokens_in, uniq_tokens_out,
                internal_tx_in, internal_tx_out,
                ci_tx_in, ci_tx_out,
                cc_tx_out,
                t_tx_in, t_tx_out,
            
                unique_in, unique_to
                

               ]

        type_vec = []
        return np.array(ls + type_vec)

        
    
    def one_hot_label(self, classes, label_switch=False):
        label_vec = [0] * len(classes)
        for index, acc in enumerate(classes):
            kw = self.label if label_switch else self.typ
            if acc == kw:
                label_vec[index] = 1
                break
        return label_vec
        
    def __str__(self):
        return f"{self.typ} {self.address}: {len(self.txs_in)} in, {len(self.txs_out)} out"

dummy = Node('', '')
feat_length = len(dummy.feature_vector(SWITCH))
feat_mask = [True] * feat_length
    
class Tx():
    
    def __init__(self, hsh, timestamp, typ, sender, receiver, contract, value, token_type=''):
        self.hsh = hsh
        self.timestamp = timestamp
        self.typ = typ
        self.sender = sender
        self.receiver = receiver
        self.value = value
        self.contract = contract
        self.token_type = token_type
        
    def __str__(self):
        return f"{self.typ}: {self.value} from {self.sender} to {self.receiver}"
    
    def __eq__(self, other):
        return  (self.__class__ == other.__class__ and
                self.hsh == other.hsh and
                self.typ == other.typ and
                self.sender == other.sender and
                self.receiver == other.receiver and
                self.value == other.value)
                
    def __hash__(self):
        return hash(('hash', self.hsh, 'type', self.typ, self.sender, self.receiver, self.value))
```

```python
def import_transactions(filename):
    transactions = []
    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0

        for row in csv_reader:

            if line_count == 0:
                line_count +=1
                continue
            else:
                transactions.append(row)
      
    return transactions

def import_exchanges(filename):
    exchanges = {}
    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            exchanges[row[0]] = ''
            
    return exchanges

def import_labelled_addresses(filename):
    addresses = {}
    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            addresses[row[0]] = [row[5], row[1], row[2], row[4]]
            
    return addresses
```

```python
def parse_external_tx(row):
    code = row[14]
    
    if code == '0x' or int(code, 0) == 0:
        tx_type = 'Transfer'     
    else:
        tx_type = 'Contract Invocation'
    
    sender, receiver, contract = row[7], row[8], row[15]
    if receiver == '':
        tx_type = 'Contract Creation'

    value = float(row[9])
    value /= 1.0E+018
#     value = 4000 if value > 4000 else value
    
    d = datetime.datetime.fromtimestamp(int(row[2]))

    tx = Tx(row[3], d, tx_type, sender, receiver, contract, value)
    return tx

def parse_internal_tx(row):
    d = datetime.datetime.fromtimestamp(int(row[2]))
    tx_type = 'Internal' 
    sender, receiver, contract = row[4], row[5], row[7]
    value = float(row[6])
    value /= 1.0E+018
#     value = 4000 if value > 4000 else value
    
    
    tx = Tx(row[3], d, tx_type, sender, receiver, contract, value) 
    return tx
    
def parse_token_tx(row):
    d = datetime.datetime.fromtimestamp(int(row[2]))
    tx_type = 'Token'
    sender, receiver, contract = row[6], row[8], row[7]
    value = 1.0E+018
    value /= 1.0E+018
#     value = 4000 if value > 4000 else value
    
    
    tx = Tx(row[3], d, tx_type, sender, receiver, contract, value, row[10])
    return tx

def parse_tx(row):  
    if row[0] == 'txlist':
        return parse_external_tx(row)
    elif row[0] == 'txlistinternal':
        return parse_internal_tx(row)
    elif row[0] == 'tokentx':
        return parse_token_tx(row)
```

```python
def gen_tx_table(tx_list):
    transactions_table = {}
    for row in tx_list:
        tx = parse_tx(row)

        if tx.hsh in transactions_table.keys():
            transactions_table[tx.hsh].add(tx)
        else:
            s = set()
            s.add(tx)
            transactions_table[tx.hsh] = s

    return transactions_table

def tx_cleanup(txs):
    senders = set()
    for tx in txs:
        if tx.typ == 'Transfer':
            senders.add(tx.sender)

    for tx in txs:
        if tx.typ == 'Contract Invocation' and tx.receiver in senders:
            tx.typ = 'Transfer'
            
    return txs

def process_transactions(tx_list):
    transactions_table = gen_tx_table(tx_list)
    
    uniq_txs = list(transactions_table.values())
    uniq_txs = [list(x) for x in uniq_txs]
    return list(itertools.chain.from_iterable(uniq_txs))

def segment_transactions(tx_list, start, end):  
    sub_list = []
    
    for tx in tx_list:
        if tx.timestamp > start and tx.timestamp < end:
            sub_list.append(tx)
            
    return sub_list

def tx_stats(transactions):
    types= {
        'Transfer': 0,
        'Contract Creation': 0,
        'Contract Invocation': 0,
        'Internal': 0,
        'Token': 0
    }
    for tx in transactions:
        types[tx.typ] +=1

    print(types)
```

```python
def assign_sender(tx_type):
    if tx_type in ['Transfer', 'Contract Invocation', 'Contract Creation']:
        return 'Wallet'
    elif tx_type in ['Internal']:
        return 'Smart Contract'
    else:
        return ''
    
def assign_receiver(tx_type):
    if tx_type in ['Transfer']:
        return 'Wallet'
    elif tx_type in ['Contract Invocation', 'Contract Creation']:
        return 'Smart Contract'
    else:
        return ''
    
#Noting internal transaction and token transfer can both have contracts as destination

def make_nodes(transactions, just_external=False):
    nodes = {}
    contracts = set()
    
    for tx in transactions:
        if just_external and tx.typ in ['Internal', 'Token']:
            continue
        
        sender, receiver = tx.sender, tx.receiver
        sender_type, receiver_type = assign_sender(tx.typ), assign_receiver(tx.typ)
                    
        if sender in nodes.keys():
            nodes[sender].add_tx_out(tx)
        else:
            n = Node(sender_type, tx.sender)
            n.add_tx_out(tx)
            nodes[sender] = n
                 
        if receiver in nodes.keys():
            nodes[receiver].add_tx_in(tx)
        else:
            n = Node(receiver_type, tx.receiver)
            n.add_tx_in(tx)
            nodes[receiver] = n
            
        if tx.contract != '':
            contracts.add(tx.contract)
            
    nodes.pop('', None) #empty node
    return nodes,contracts

def label_contracts(nodes, contracts):
    
    for addr in nodes.keys():
        if nodes[addr].typ == '' and nodes[addr].address in contracts:
            nodes[addr].typ = 'Smart Contract'
```

```python
def label_exchanges(node_dict, exchange_list):
    
    for addr in node_dict.keys():
        if node_dict[addr].address in exchange_list.keys():
            node_dict[addr].typ = 'Exchange'
            
def label_entities(node_dict, labelled_addr):
    for addr in node_dict.keys():
        if node_dict[addr].address in labelled_addr.keys():
            e = labelled_addr[addr][3]  
            node_dict[addr].label = e

def label(node_dict, labelled_addrs):
    for addr in node_dict.keys():
        if node_dict[addr].address in labelled_addr.keys():
            l = 'Fraudulent' if labelled_addr[addr][0] == 'Dodgy' else 'Verified'
            node_dict[addr].label = l
            node_dict[addr].name = labelled_addr[addr][1]
            node_dict[addr].typ = labelled_addr[addr][2]
        else:
            p = randint(0,1000)
            if p > 995:
                node_dict[addr].label = 'z'
                node_dict[addr].name = addr
        
def accounts_stats(node_dict, label_switch = False):
    acc_dist = {}
    
    for addr, node in node_dict.items():
        kw = node.label if label_switch else node.typ
        if kw in acc_dist.keys():
            acc_dist[kw] += 1
        else:
            acc_dist[kw] = 0
            
    print(acc_dist)
    classes = sorted(list(acc_dist.keys()))
    if 'Wallet' in classes:
        classes.reverse()
    return acc_dist, classes
```

```python
def nodes_preprocess(nodes, classes, feats_mask, label_switch):
    l = len(nodes)
    feat_length = sum(feats_mask)
    classes_length = len(classes)
    
    addresses = np.zeros((l), dtype=object)
    features = np.zeros((l, feat_length), dtype='float32')
    labels = np.zeros((l, classes_length), int)
    
    for i, (addr, node) in enumerate(nodes.items()):
        addresses[i] = addr
        feats = np.array(node.feature_vector(label_switch))
        features[i] = feats[feats_mask]
        labels[i] = node.one_hot_label(classes, label_switch)
    
    return addresses, features, labels
```

```python
def shuffle(addresses, features, labels):
    assert len(labels) == len(addresses) == len(features)
    p = np.random.permutation(len(labels))
    labels = labels[p]
    addresses = addresses[p]
    features = features[p]
    
    return addresses, features, labels
```

```python
def normalize(vec):
    mins = vec.min(axis=0)
    maxs = vec.max(axis=0)
    return (vec - mins) / (maxs - mins)

def log_normalize(vec):
    return np.where(vec > 0, np.log(vec), 0)
```

```python
def assign_node_ids(addresses, nodes):

    for i in range(len(addresses)):
        nodes[addresses[i]].node_id = i
    
    return nodes

def gen_labelled_mask(addresses, nodes, label_switch=False ):
    if label_switch:
        return [nodes[x].label != '' for x in addresses]
    else:
        return [nodes[x].typ != '' for x in addresses]

```

```python
def build_edges(transactions, node_dict):
    l = len(transactions)
    u = np.zeros((l), int)
    v = np.zeros((l), int)
    e = np.zeros((l), float)
    
    for i, tx in enumerate(transactions):
        if tx.sender in node_dict.keys() and tx.receiver in node_dict.keys():
            u[i] = node_dict[tx.sender].node_id
            v[i] = node_dict[tx.receiver].node_id
            e[i] = float(tx.value)
            
    return u, v, e

def build_edges_undirected(transactions, node_dict):
    l = len(transactions)
    src = np.zeros((l), int)
    dest = np.zeros((l), int)
    e = np.zeros((l), float)
    
    for i, tx in enumerate(transactions):
        if tx.sender in node_dict.keys() and tx.receiver in node_dict.keys():
            src[i] = node_dict[tx.sender].node_id
            dest[i] = node_dict[tx.receiver].node_id
            e[i] = float(tx.value)
    
    u = np.concatenate([src, dest])
    v = np.concatenate([dest, src])
    e = np.concatenate([e,e])
    
    return u, v, e

def transform_edges(e):
    e = normalize(e)
    
    e = np.float32(e)
    e = torch.from_numpy(e)
    e = e.to(torch.device('cuda:0'))
    #pad the edges features with 0's for the self loops
    e = nn.ConstantPad1d((0, len(g.edges) - len(e)), 1)(e)
    
    return e
```

```python
def to_tensor(arr, float32=False):
    if float32:
        arr = np.float32(arr)
    
    arr = torch.from_numpy(arr)
    return arr

def to_gpu(tens):
    tens = tens.to(torch.device('cuda:0'))
    return tens
```

```python
def class_weights(acc_dist):
    total = 0
    acc_dist.pop('')
    for x in acc_dist.values():
        total += x
    
    for acc in acc_dist.keys():

        acc_dist[acc] /= total
        acc_dist[acc] = 1 - acc_dist[acc]
    

    classes_w = np.array(list(acc_dist.values()))
    
    return classes_w
```

```python
def train_test_mask(length):

    train_mask = np.empty((length), bool)

    mask_border = int(length // 1.25) #80% train, 20% test

    for i in range(length):
        train_mask[i] = True if i < mask_border else False

    test_mask = np.invert(train_mask) 
    
    return train_mask, test_mask
```

```python
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=14)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, fontsize=12)
    plt.yticks(tick_marks, classes, fontsize=12)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black", fontsize=14)

    plt.tight_layout()
    plt.ylabel('True label',fontsize=14)
    plt.xlabel('Predicted label', fontsize=14)
    
def accuracy(matrix):
    total = 0
    for x in matrix:
        total += sum(x)
        
    correct = 0
    for i in range(len(matrix)):
        correct += matrix[i][i]
        
    return correct / total

def precision(matrix, j):
    col_sum = 0
    for i in range(len(matrix)):
        col_sum += matrix[i][j]
        
    return 0 if col_sum == 0 else matrix[j][j] / col_sum

def recall(matrix, i):
    
    row_sum = sum(matrix[i])
    
    return 0 if row_sum == 0 else matrix[i][i] / row_sum

def mac_avg_metric(matrix, metric):
    res = []
    
    for i in range(len(matrix)):
        res.append(metric(matrix, i))
        
    return sum(res) / len(res)

def f1_score(precision, recall):
    
    return 2* ((precision * recall) / (precision + recall))
```

```python
class GCNLayer(nn.Module):
    
    def __init__(self, in_feats, out_feats, msg_func, feat_pad = 0):
        super(GCNLayer, self).__init__()
        self.msg_func = msg_func
        self.linear = nn.Linear(in_feats + feat_pad, out_feats)  
        
    def forward(self, g, feature):
        with g.local_scope():
            g.ndata['h'] = feature
            g.update_all(self.msg_func, gcn_reduce)
            h = g.ndata['h']
            return self.linear(h)
```

```python
def mid_point(in_feats, classes):
    return ((in_feats - classes) // 2) + classes

class Net(nn.Module):
    def __init__(self, msg_func, pad=0):
        super(Net, self).__init__()
        mid = mid_point(feat_length, classes_length)
        self.layer1 = GCNLayer(feat_length, feat_length, msg_func, pad)
        self.layer2 = GCNLayer(feat_length, mid, msg_func, pad)
        
#         self.layer1 = nn.Linear(feat_length, feat_length)
#         self.layer2 = nn.Linear(feat_length, mid)
        
        self.layer3 = nn.Linear(mid, classes_length)

  
    def forward(self, g, features):
        x = torch.relu(self.layer1(g, features))
        x = torch.relu(self.layer2(g, x))
        
#         x = torch.relu(self.layer1(features))
#         x = torch.relu(self.layer2(x))
        
        x = torch.relu(self.layer3(x))
#         x = F.relu(self.layer4(x))
        
        return x
#         return F.softmax(x, 1)
```

```python
def transactions_to_edges(txs, node_dict):
    l = len(txs)
    src = np.zeros((l), int)
    dest = np.zeros((l), int)
    e = np.zeros((l), dtype='float')
    
    for i, tx in enumerate(transactions):
        if tx.sender in node_dict.keys() and tx.receiver in node_dict.keys():
            src[i] = node_dict[tx.sender].node_id
            dest[i] = node_dict[tx.receiver].node_id
            e[i] = float(tx.value)
    
    u = np.concatenate([src, dest])
    v = np.concatenate([dest, src])
    e = np.concatenate([e,e])
#     u, v, e = src, dest, e
    ones = np.ones((len(nodes)))
    e = np.concatenate([e, ones])
    
    return to_tensor(u), to_tensor(v), to_tensor(e)

def txs_to_pyg(u, v, e, feats):
    
    edge_index = torch.stack((u.t(), v.t()), dim=0)
    edge_index = edge_index.squeeze()
    
    g = Data(x=feats, edge_index=edge_index, edge_attr=e)
    
    return g
```

```python
from torch_geometric.nn import MessagePassing
from torch_geometric.utils import add_self_loops, degree

def mid_point(in_feats, classes):
    return ((in_feats - classes) // 2) + classes

class GCNConv(MessagePassing):
    def __init__(self, in_channels, out_channels, e):
        super(GCNConv, self).__init__(aggr='add')  # "Add" aggregation (Step 5).
        self.lin = torch.nn.Linear(in_channels, out_channels)
        self.e = e

    def forward(self, x, edge_index):
        # x has shape [N, in_channels]
        # edge_index has shape [2, E]

        # Step 1: Add self-loops to the adjacency matrix.
        edge_index, _ = add_self_loops(edge_index, num_nodes=x.size(0))

        # Step 2: Linearly transform node feature matrix.
        x = self.lin(x)

        # Step 3: Compute normalization.
        row, col = edge_index
        deg = degree(col, x.size(0), dtype=x.dtype)
        deg_inv_sqrt = deg.pow(-0.5)
        norm = deg_inv_sqrt[row] * deg_inv_sqrt[col]

        # Step 4-5: Start propagating messages.
        return self.propagate(edge_index, x=x, norm=norm, e=self.e)

    def message(self, x_j, norm):
        # x_j has shape [E, out_channels]

        # Step 4: Normalize node features.
        b = (norm).view(-1, 1) * x_j 
        return b 

class NetPyG(torch.nn.Module):
    def __init__(self, num_features, num_classes, e):
        super(NetPyG, self).__init__()
        mid = mid_point(feat_length, classes_length)
        self.conv1 = GCNConv(num_features, num_features, e)
        self.conv2 = GCNConv(num_features, mid, e)
        self.layer3 = nn.Linear(mid, num_classes)

    def forward(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        x = F.dropout(x, training=self.training)
        x = x.to(torch.float32)
        x = self.conv2(x, edge_index)
        x = x.to(torch.float32)
        x = torch.relu(self.layer3(x))
        return F.log_softmax(x, dim=1)
```
